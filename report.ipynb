# Končno poročilo o analizi zasedenosti parkirnih mest v Ljubljani in okolici

Naredila sva model, ki z uporabo časovnih vrst v zadovoljivi meri napoveduje zasedenost posameznih parkirišč ter spletno aplikacijo, ki uporablja ta model za priporočanje optimalnega parkirnega mesta, glede na izbrano lokacijo in čas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from foursquareAPI import get_poi
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAXResults
import statsmodels.api as sm
from pandas.tseries.offsets import DateOffset
from math import sqrt
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Parkirisca_do_10_05_2022.csv") # TODO: Use cols here
df["Zasedenost"] = df["Kapaciteta"] - df["Prosta mesta"]
df['date'] = pd.to_datetime(df["Datum"], unit="s")
df = df.drop(["Datum", "Prosta (narocniki)", "Oddana (narocniki)",
             "Na voljo (narocniki)", "Cakalna vrsta (narocniki)"], axis=1)
df
```

## Izris (pomankljivih) podatkov

Spodnji graf prikazuje zasedenost enega od parkirišč (BS4/ I) skozi čas. Kot je iz grafa razvidno, so med podatki "luknje", saj se je skripta za "scrapanje" podatkov 2-krat sesula. To pomeni, da bomo podatke morali **imputirati**.

```python
name, parking = next(iter(df.groupby("Parkirisce")))
name = str(name)

parking = parking.set_index('date')
parking = parking.drop(["Parkirisce", "Prosta mesta"], axis=1)

plt.figure(figsize=(12, 6))
plt.gcf().autofmt_xdate()
plt.xlabel("Datum")
plt.ylabel("Stevilo zasedenih mest")
plt.title(f"Parkirisce: {name}")
plt.plot(parking["Zasedenost"]);
```

## Preprocesiranje podatkov

### Imputacija
Spodnja koda imputira podatke za določeno parkirišče tako, da ponovi zasedenost iz prejšnjega tedna na isti dan. Poleg tega pa še sinhronizira frekvenco časovnih intervalov na 10 minut, kar je ključnega pomena, če želimo napovedovati podatke z uporabo časovnih vrst.

```python
def copy_and_impute(parking):
    # Prepare new data frame, which will be synced (freqeuncy is exactly 10 mins)
    parking_sync = parking.asfreq("10 min")

    # Copy data before hole 1
    hole_start_date1 = pd.to_datetime("2022-03-15 08:25:12", format='%Y-%m-%d %H:%M:%S')
    for ((index, row), (index_sync, row_sync)) in zip(parking.iterrows(), parking_sync.iterrows()):
        if index > hole_start_date1:
            break
        row_sync["Zasedenost"] = row["Zasedenost"]

    # Impute first hole
    hole_end_date1 = pd.to_datetime("2022-03-23 07:02:59", format='%Y-%m-%d %H:%M:%S')
    num_rows_imputed = 0
    for index_sync, row_sync in parking_sync.iterrows():
        if index_sync <= hole_start_date1:
            continue
        if index_sync >= hole_end_date1:
            break
        # Repeat data from previous week day
        row_sync["Zasedenost"] = parking_sync[parking_sync.index ==
                                              index_sync - pd.to_timedelta(1, unit="W")]["Zasedenost"][0]
        num_rows_imputed += 1

    # Copy data after hole 1, before hole 2
    hole_start_date2 = pd.to_datetime("2022-04-07 22:10:55", format='%Y-%m-%d %H:%M:%S')
    for ((index, row), (index_sync, row_sync)) in zip(parking.iterrows(), parking_sync.iloc[num_rows_imputed:].iterrows()):
        if index < hole_end_date1:
            continue
        if index >= hole_start_date2:
            break
        row_sync["Zasedenost"] = row["Zasedenost"]

    # Impute second hole
    hole_end_date2 = pd.to_datetime("2022-04-22 08:11:54", format='%Y-%m-%d %H:%M:%S')
    for index_sync, row_sync in parking_sync.iterrows():
        if index_sync <= hole_start_date2:
            continue
        if index_sync > hole_end_date2:
            break
        # Repeat data from previous week day
        row_sync["Zasedenost"] = parking_sync[parking_sync.index ==
                                              index_sync - pd.to_timedelta(1, unit="W")]["Zasedenost"][0]
        num_rows_imputed += 1

    # Copy the remaning data
    hole_start_date2 = pd.to_datetime("2022-04-07 22:10:55", format='%Y-%m-%d %H:%M:%S')
    for ((index, row), (index_sync, row_sync)) in zip(parking.iterrows(), parking_sync.iloc[num_rows_imputed:].iterrows()):
        if index < hole_end_date2:
            continue
        row_sync["Zasedenost"] = row["Zasedenost"]

    parking_sync = parking_sync["Zasedenost"].fillna(method="bfill")  # Fill remaning spaces with previous values
    parking_sync.dropna(axis=0, inplace=True)  # Remove the somehow remaning NaN rows

    return parking_sync
```

### Dodajanje atributov, filtriranje napačnih podatkov

Med atribute sva vključila tudi podatek o lokaciji, vendar, ker lokacija sama po sebi modelu ne bi pomagala, sva vključila **razdalje do bližnjih točk interesa** (restavracije, muzeji in bari). Vrednost vsakega atributa je: $ max\ razdalja + 1 - razdalja $, vrednost 0 torej pomeni, da v okolici ni točke interesa, vrednosti večje od 0 pa predstavljajo razliko razdalje do točke interesa. Tako večja vrednost pomeni "boljše" parkirišče, in bo model predvidoma lažje ekstrapoliral informacije iz tega atributa.

```python
parkings = {}
for name, group in df.groupby("Parkirisce"):
    if name in ("BS4/ I", "BS4/ II"): # Bad data
        continue    
    name = str(name)
    parking = group

    parking = parking.set_index('date')
    parking = parking.drop(["Parkirisce", "Prosta mesta"], axis=1)

    capacity = 0
    for _, row in parking.iterrows():
        if row["Kapaciteta"] != 0:
            capacity = row["Kapaciteta"]
            break

    raidus = 50
    points_of_interest = get_poi(name, radius=raidus)

    # Filter empty data (10 parkings)
    if (parking["Zasedenost"] == 0).all() or (parking["Zasedenost"] == capacity).all() or points_of_interest is None:
        continue

    parking_sync = pd.DataFrame(copy_and_impute(parking))
    parking_sync["Kapaciteta"] = [capacity for _ in range(len(parking_sync.index))]
    parking_sync["Procent_zasedenosti"] = parking_sync["Zasedenost"] / parking_sync["Kapaciteta"]

    # Add points of interest to the specific column (0: not present, radius-distance: point of interest score) -> POIs are weighted
    restaurant_dist = []
    art_dist = []
    coffee_dist = []
    bar_dist = []
    for distance, poi in points_of_interest:
        if "Restaurant" in poi:
            restaurant_dist.append(distance)
        elif "Art" in poi or "Museum" in poi:
            art_dist.append(distance)
        elif "Coffee" in poi or "Café" in poi or "Cafeteria" in poi:
            coffee_dist.append(distance)
        elif "Bar" in poi:
            bar_dist.append(distance)
    
    restaurant_score = 0 if not restaurant_dist else raidus + 1 - sum(restaurant_dist) / len(restaurant_dist)
    art_score = 0 if not art_dist else raidus + 1 - sum(art_dist) / len(art_dist)
    coffee_score = 0 if not coffee_dist else raidus + 1 - sum(coffee_dist) / len(coffee_dist)
    bar_score = 0 if not bar_dist else raidus + 1 - sum(bar_dist) / len(bar_dist)
    
    parking_sync["Restaurant_score"] = [restaurant_score for _ in range(len(parking_sync.index))]
    parking_sync["Art_score"] = [art_score for _ in range(len(parking_sync.index))]
    parking_sync["Coffee_score"] = [coffee_score for _ in range(len(parking_sync.index))]
    parking_sync["Bar_score"] = [bar_score for _ in range(len(parking_sync.index))]

    parking_sync = parking_sync.astype({"Zasedenost": int, "Kapaciteta": int, "Procent_zasedenosti": float,
                                               "Restaurant_score": float,	"Art_score": float,	"Coffee_score": float,	"Bar_score": float})

    parkings[name] = parking_sync  # Save the synced data frame to dict

print(parkings)
```

```python
name = next(iter(parkings.keys()))
parking = parkings[name]

plt.figure(figsize=(12, 6))
plt.gcf().autofmt_xdate()
plt.xlabel("Datum")
plt.ylabel("Stevilo zasedenih mest")
plt.title(f"Parkirisce: {name}")
plt.plot(parking["Zasedenost"]);
```

## Dekompozicija in preoblikovanje podatkov

Spodnji graf prikazuje "sezonskost" podatkov. Kot je to iz grafa razvidno, seznonskosti ni bilo mogoče razpoznati, zato je podatke potrebno nekoliko preobilikovati. To je storjeno tako, da so podatki v istem dnevu združeni in povprečeni.

```python
decompfreq = 1 * 60 // 10 # seasonality: 1/h, frequency: 10 min
results = seasonal_decompose(parkings[name]["Zasedenost"],period=decompfreq)
results.plot();
```

Spodnji graf prikazuje povpreče posameznega dneva, tako lahko model vidi "širšo sliko", nato pa je spet prikazana veliko boljša dekompozicija. Iz dekompozijce je tokrat možno razbrati trend in sezonskost, tudi rezidualni podatki (ostanek) izgledajo naključni, kar pomeni, da smo iz podatkov ekstrapolirali vse informacije.

```python
curr_df = parkings[name][["Zasedenost", "Restaurant_score", "Art_score", "Coffee_score", "Bar_score"]]
data = curr_df.resample("D").mean()

plt.figure(figsize=(12, 6))
plt.gcf().autofmt_xdate()
plt.xlabel("Datum")
plt.ylabel("Stevilo zasedenih mest")
plt.title(f"Parkirisce: {name}")
plt.plot(data["Zasedenost"])

results = seasonal_decompose(data["Zasedenost"])
results.plot();
```

Podatki so sedaj pripravljeni za učenje modela. Na tej točki bi lahko uporabili tudi `adfuller` test, ki preveri ali so podatki stacionarni, vendar je uporabljen model `SARIMAX` (Seasonal Auto-Regressive Integrated Moving Average with eXogenous factors), pri katerem stacionarnost ni pomembna, saj upošteva tudi sezonskost.

Spodaj je ustvarjen model, ki upošteva tudi razdalje do točk interesa oziroma z "exogenous" faktorji, nato pa je prikazan graf predikcije modela. Čeprav je časovni interval za predikcijo znoraj učne množice to ni pomembno, saj model generalizira učne podatke in jih nato ne upošteva več.

```python
model = sm.tsa.statespace.SARIMAX(data['Zasedenost'], exog=data[[
                                  "Restaurant_score", "Art_score", "Coffee_score", "Bar_score"]], order=(6, 1, 4), seasonal_order=(1, 1, 1, 52)).fit()

data["Predikcija_exog"] = model.predict(start='2022-05-02', end='2022-05-10', dynamic=True)
data[['Zasedenost', 'Predikcija_exog']].plot(figsize=(12, 8))
```

Kot je to razvidno iz grafa, je model podatke napovedal kar dobro. Spodaj je izpisan RMSE modela.

```python
rmse = sqrt(mean_squared_error(data['Zasedenost'][-9:], data['Predikcija_exog'][-9:]))
print("RMSE z exogenous faktorji:", rmse)
```

V naslednjem odseku je ustvarjen model, ki ne upošteva lokacij to točk interesa, vendar le vrednosti zasedenosti.

```python
model=sm.tsa.statespace.SARIMAX(data['Zasedenost'],order=(6, 1, 4),seasonal_order=(1,1,1,52)).fit()

data['Predikcija'] = model.predict(start='2022-05-02', end='2022-05-10', dynamic=True)
data[['Zasedenost', 'Predikcija']].plot(figsize=(12, 8))
```

```python
rmse = sqrt(mean_squared_error(data['Zasedenost'][-9:], data['Predikcija'][-9:]))
print("RMSE brez exogenous faktorjev:", rmse)
```

Kot vidimo ima model, ki ne upošteva razdalj celo **manjši** RMSE. Ker je razlika med tema dvema modeloma zanemarljiva, bo za priporočilni sistem uporabljen slednji, saj je preprostejši.
